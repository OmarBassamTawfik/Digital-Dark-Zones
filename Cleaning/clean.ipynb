```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_grouped = df.groupby('quadkey', as_index=False)[numeric_columns].mean()
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# Create a GeoDataFrame for the points
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
gdf_points = gpd.GeoDataFrame(data, geometry=geometry, crs="EPSG:4326")  # Assuming WGS84 (lat/lon)

# Load Brazil's border shapefile/GeoJSON
brazil_border = gpd.read_file("../110m_cultural/ne_110m_admin_0_countries.shp")  # Update the path
brazil_border = brazil_border.to_crs("EPSG:4326")  # Ensure the same CRS

# Filter to keep only Brazil's border
brazil_border = brazil_border[brazil_border['ADMIN'] == 'Brazil']

# Perform spatial join to keep only points inside Brazil
filtered_gdf = gdf_points[gdf_points.geometry.within(brazil_border.unary_union)]

# Convert back to a pandas DataFrame if needed
filtered_data = filtered_gdf.drop(columns='geometry')

# Display the info of the original data
print("Original data info:")
data.info()

# Display the info of the filtered data
print("Filtered data info:")
filtered_data.info()

# Save the filtered data to a new CSV file
filtered_data.to_csv("Brazil_Data.csv", index=False)

# Display the filtered data
print(filtered_data)
```
